```python
import pandas as pd 
import numpy as np
from pandas import NA
df = pd.DataFrame(np.random.randn(7, 3))
df
```

```python
df.iloc[:4, 1] = NA
df.iloc[:2, 2] = NA

# 每行NA数量大于2
df.dropna(thresh=2)
```

```python
df.fillna(0)

# 指定索引默认值
df.fillna({1: 0.5, 2: 0.1})
```

```python
df = pd.DataFrame(np.random.randn(6, 3))
df.iloc[2:, 1] = NA
df.iloc[4:, 2] = NA
# 以当列的最后一个值作为默认值
df.fillna(method='ffill')

```

```python
# map
data = pd.DataFrame({'food': ['bacon', 'pulled pork', 'bacon',
    'Pastrami', 'corned beef', 'Bacon',
   'pastrami', 'honey ham', 'nova lox'],
   'ounces': [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})

meat_to_animal = {
  'bacon': 'pig',
  'pulled pork': 'pig',
  'pastrami': 'cow',
  'corned beef': 'cow',
  'honey ham': 'pig',
  'nova lox': 'salmon'
}

lowercased = data['food'].str.lower()
data['animal'] = lowercased.map(meat_to_animal)
data['food'].map(lambda x: meat_to_animal[x.lower()])
```

```python
# replace
data = pd.Series([1., -999., 2., -999., -1000., 3.])
data.replace(-999, 0)

data.replace([-999, 2], 0)

data.replace([-999, 2], [-998, 4.])
data.replace({-999: 0, 2: 4})
```
